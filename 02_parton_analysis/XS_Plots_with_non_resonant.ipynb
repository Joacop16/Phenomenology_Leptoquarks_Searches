```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
M =[int(M)/1000 for M in ['0250', '0500', '0750','1000', '1250', '1500', '1750', '2000', '2250', '2500']]
```

```python
XS_LQ_Tau = np.array([182.3, 5.944, 0.6093, 0.1029, 0.02291, 0.006099, 0.001836, 0.0006038, 0.0002119, 7.806e-5])
XS_Tau_Tau = np.array([40.33, 4.184, 9.95e-1, 3.45e-1, 1.47e-1, 7.29e-2, 3.96e-2, 2.337e-2, 1.47e-2, 9.68e-3])
XS_LQ_LQ = np.array([937.9, 20.02, 1.527, 1.957e-1, 3.301e-2, 6.541e-3, 1.428e-3, 3.319e-4, 8.022e-5, 1.997e-5])
```

```python
fig, ax = plt.subplots(figsize = (5,5))

plt.plot(M, XS_LQ_Tau, label = 'sLQ')
plt.plot(M, XS_LQ_LQ, label = 'dLQ')
plt.plot(M, XS_Tau_Tau, label = 'non-resonant')
plt.plot(M, XS_Tau_Tau +  XS_LQ_LQ + XS_LQ_Tau, label = 'sum')

plt.legend()
plt.grid(True, which="both", ls="-")
plt.xlabel(r'$M_{U}$ [TeV]', loc= 'right', fontsize = 12)
plt.ylabel(r'$\sigma $ (pb)', loc = 'top', fontsize = 12)
plt.xlim(np.min(M), np.max(M))

plt.xticks(np.arange(0.25, 2.5, step=0.2))

plt.yscale('log')
ax.set_yticks([10**-5, 10**-4, 10**-3, 10**-2, 10**-1, 10**0, 10**1, 10**2, 10**3])

plt.title(r'$\mathbf{\sqrt{s} = 13 TeV, g_{U} = 1.8}$', fontweight='bold', loc = 'right')
plt.show()
```
